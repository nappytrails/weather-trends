```python
# Dependencies
import json
import requests 
import pandas as pd
from time import ctime
from pprint import pprint
```

```python
# Specify the URL
# Major cities in California

forecastCity = "Los Angeles"
dailyURL = "https://api.weather.gov/gridpoints/LOX/154,44/forecast"
hourlyURL = "https://api.weather.gov/gridpoints/LOX/154,44/forecast/hourly"
```

```python
# Make request and store response
dailyResponse = requests.get(dailyURL).json()
pprint(dailyResponse)
```

```python
dailyForecasts = dailyResponse['properties']['periods']

for dailyForecast in dailyForecasts:
    pprint(dailyForecast)
```

```python
dailyForecast = dailyResponse['properties']['periods'][0]
# pprint(dailyForecast)
dailyForecast['city']= forecastCity
dailyForecast['runTime']= [ctime()]
dailyForecast
```

```python
# Make request and store response
hourlyResponse = requests.get(hourlyURL).json()
pprint(hourlyResponse)
```

```python
hourlyForecast = hourlyResponse['properties']['periods'][0] 
# pprint(hourlyForecast)
hourlyForecast['city']= forecastCity
hourlyForecast['runTime']= [ctime()]
hourlyForecast
```

```python
dailyForecastDF = pd.DataFrame.from_dict(dailyForecast)
dailyForecastDF = dailyForecastDF.rename(columns = {'number':"responseNumber",
                                                    'name':"responseName"})
dailyForecastDF
```

```python
hourlyForecastDF = pd.DataFrame.from_dict(hourlyForecast)
hourlyForecastDF = hourlyForecastDF.rename(columns = {'number':"responseNumber",
                                                    'name':"responseName"})
hourlyForecastDF
```

```python
from config import driver, username, password, host, port, database
from sqlalchemy import create_engine
```

```python
connection_string = f"{driver}://{username}:{password}@{host}:{port}/{database}"
engine = create_engine(connection_string)
connection = engine.connect()
```

```python
dailyForecastDF.to_sql('dailyForecastTB',connection, if_exists='append', index=False)
dailyForecastDF.head()
```
